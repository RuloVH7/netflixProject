# TASK 3: In which year were the highest number of TV Shows and Movies releases? Show the results with a bar graph.

```python
# For this task it is needed to work with dates, so, a conversion from object to datetime data type.

# pd.to_datetime() function is incredibly useful when exist dates or times in various formats (e.g., strings or 
# numeric values) and are needed to convert into pandas datetime objects to perform various date and time 
# operations and manipulations.

# Result will be stored on a new column named ['Date_N'].
netflix_data['Date_N'] = pd.to_datetime(netflix_data['Release_Date'], yearfirst=True)
```

```python
# Check the dataframe with the new column
netflix_data.head()
```

```python
# Confirm the data types.
netflix_data.dtypes
```

```python
# Next step consist on counting the occurrence of all individuals years in date column.

# dt.year attribute specifically allows accessing the year component of the datetime objects in the 
# Series or DataFrame column.

# Use value_counts() to count how many times each year are repeated.
netflix_data['Date_N'].dt.year.value_counts()
```

```python
# sns.countplot() function is specifically designed to create bar plots for categorical data and is used to 
# visualize the distribution of observations within a single categorical variable. 

# sns.set_context() function is used to set the context for the Seaborn plots, which controls the overall 
# appearance of the visualizations.

bar_graph = sns.countplot(data=netflix_data, x=netflix_data['Date_N'].dt.year, hue='Category', palette='Set1')
sns.set_context("poster", font_scale = 0.3, rc={"grid.linewidth": 0.6})
bar_graph.set_title('TV-Shows and Movies per year')
bar_graph.set_xlabel('Year')
bar_graph.set_ylabel('# of titles')
for label in bar_graph.containers:
    bar_graph.bar_label(label)
plt.show()

```

# TASK 10: What is the maximun duration of a Movie/ TV Show on Netflix?

```python
# Calling the dataframe.
netflix_data.head()
```

```python
# It is showing 2 different measurements. One is minutes (For movie) and Seasons (for tv-shows).

# str.split() unction is particularly useful to separate the strings into multiple parts or extract specific 
# information from a column. For example to split a comma-separated string into individual items.

# expand: If True, the result will be returned as a DataFrame with each split part in a separate column. If 
# False (default), the result will be returned as a Series with lists of split parts. For this expand will be
# true to create 2 columns: Quantity and Unit
netflix_data[['Quantity', 'Unit']] = netflix_data['Duration'].str.split(' ', expand=True)
```

```python
# Check for the result
netflix_data.head()
```

```python
# Second step will be to convert the quantity values to integers (int) to apply the max() and min() function.

# astype() function convert the data type of elements to another type, such as changing numerical values to 
# strings or strings to integers.

# netflix_data.Quantity is the same as netflix_data['Quantity']
netflix_data.Quantity = netflix_data.Quantity.astype(int)
```

```python
# Another issue arises. max() and min() function will look just numbers, but there are 2 different unist of 
# measurement.
# To ensure that the methods will work for each category movie and tv-show. Filtering season and min and saved 
# them into a separated dataframe resepectively will be the next step.
netflix_seasons = netflix_data[netflix_data['Unit'] == 'Seasons']
netflix_minutes = netflix_data[netflix_data['Unit'] == 'min']
```

```python
# Finally, looking into 'Quantity' column from seasons, the maximun value can be found.
netflix_seasons['Quantity'].max()
```

```python
# Then, in minutes dataframe
netflix_minutes.Quantity.max()
```
